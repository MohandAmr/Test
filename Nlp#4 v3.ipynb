```python
df["target"]=df["target"].map({4:1,0:0})
df["target"].value_counts()
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https?://\S+|#\S+|@\S+|[0-9]+|\W", " ", text)
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return " ".join(filtered_tokens)

df['cleaned_text'] = df['text'].apply(preprocess_text)
df.head()
```

```python
# Parameters
vocab_size = 10000  # maximum number of words to keep, based on word frequency
max_length = 100    # maximum length of all sequences
embedding_dim = 100 # embedding dimensions for word vectors

# Tokenization
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(df['cleaned_text'])
sequences = tokenizer.texts_to_sequences(df['cleaned_text'])

# Padding
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post', truncating='post')

# Assuming the target column is binary and named 'target'
y = np.array(df['target'])
```

```python
def create_cnn_model(vocab_size, embedding_dim, max_length, num_filters, kernel_size, num_conv_layers, num_dense_units, dropout_rate, num_classes):
    model = Sequential([
        Embedding(vocab_size, embedding_dim, input_length=max_length),
        Conv1D(num_filters, kernel_size, activation='relu')
    ])

    # Add additional convolutional layers if requested
    for _ in range(1, num_conv_layers):
        model.add(Conv1D(num_filters, kernel_size, activation='relu'))
    
    model.add(GlobalMaxPooling1D())
    model.add(Dense(num_dense_units, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(num_classes, activation='softmax'))
    
    return model
```

```python
def compile_and_train_model(model, X_train, y_train, optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'], epochs=10, batch_size=32, validation_split=0.2):
    model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, validation_split=validation_split, verbose=2)
    return history
```

```python
def plot_model_architecture(model, filename='model.png'):
    """
    Plots the architecture of a Keras model.

    Args:
    model: Keras model instance to be plotted.
    filename: Name of the file to save the plot. The plot is also displayed using matplotlib.

    Returns:
    None
    """
    # Generate the plot
    plot_model(model, to_file=filename, show_shapes=True, show_layer_names=True)
    
    # Display the plot
    image = plt.imread(filename)
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
# Define the parameters for the five models
model_params = [
    {'num_filters': 32, 'kernel_size': 3, 'num_conv_layers': 1, 'num_dense_units': 64, 'dropout_rate': 0.2},
    {'num_filters': 64, 'kernel_size': 3, 'num_conv_layers': 2, 'num_dense_units': 128, 'dropout_rate': 0.3},
    {'num_filters': 128, 'kernel_size': 5, 'num_conv_layers': 2, 'num_dense_units': 128, 'dropout_rate': 0.4},
    {'num_filters': 64, 'kernel_size': 3, 'num_conv_layers': 3, 'num_dense_units': 256, 'dropout_rate': 0.5},
    {'num_filters': 128, 'kernel_size': 3, 'num_conv_layers': 1, 'num_dense_units': 64, 'dropout_rate': 0.2}
]

models = []
for params in model_params:
    model = create_cnn_model(
        vocab_size=10000, 
        embedding_dim=100, 
        max_length=100, 
        num_filters=params['num_filters'], 
        kernel_size=params['kernel_size'], 
        num_conv_layers=params['num_conv_layers'], 
        num_dense_units=params['num_dense_units'], 
        dropout_rate=params['dropout_rate'], 
        num_classes=2
    )
    models.append(model)
```

```python
def train_models(models, X_train, y_train):
    histories = []
    for model in models:
        history = compile_and_train_model(
            model, 
            X_train, 
            y_train, 
            epochs=10, 
            batch_size=64
        )
        histories.append(history)
    return histories

# Assuming you have padded_sequences as X_train and y as y_train
histories = train_models(models, padded_sequences, y)
```

```python
for i, model in enumerate(models):
    plot_model_architecture(model, filename=f'model_{i+1}.png')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for i, history in enumerate(histories):
    plt.plot(history.history['accuracy'], label=f'Model {i+1} Train Accuracy')
    plt.plot(history.history['val_accuracy'], label=f'Model {i+1} Val Accuracy')

plt.title('Comparison of CNN Models')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
